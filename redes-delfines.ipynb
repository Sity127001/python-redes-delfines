Importar los datos:

Tenemos:

- 8 ficheros de nodos
- 8 ficheros de aristas

Uno por temporada (s1, s2, …)

Vamos a recorrer las temporadas con un for

Construir un grafo por temporada

Guardarlos en un diccionario: graphs["s1"], graphs["s2"], etc.

Además:

Si en edges aparece un nodo que no está en nodes --> NetworkX lo añade con **add_edge**.

```python
import pandas as pd
import networkx as nx

data_path = "data"

graphs = {}  # un grafo por temporada

for season in range(1, 9):
    # Rutas a los ficheros
    nodes_file = f"{data_path}/got-s{season}-nodes.csv"
    edges_file = f"{data_path}/got-s{season}-edges.csv"

    # Cargamos CSVs
    nodes_df = pd.read_csv(nodes_file)
    edges_df = pd.read_csv(edges_file)

    # Creamos grafo
    G = nx.Graph(season=season)

    # Añadimos nodos
    for _, row in nodes_df.iterrows():
        G.add_node(row["Id"], label=row["Label"])

    # Añadimos aristas
    for _, row in edges_df.iterrows():
        G.add_edge(
            row["Source"],
            row["Target"],
            weight=row["Weight"],
            season=row["Season"]
        )

    # Guardamos el grafo
    graphs[f"s{season}"] = G

    # Info solicitada
    print(f"Temporada {season}:")
    print("  Nodos:", G.number_of_nodes())
    print("  Aristas:", G.number_of_edges())
    print("-" * 30)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

# Layout base usando la temporada 1
base_G = graphs["s1"]
base_pos = nx.spring_layout(base_G, seed=42)

fig, ax = plt.subplots(figsize=(14, 6))

layer_spacing = 3  # distancia entre capas

for i, (season_key, G) in enumerate(graphs.items()):
    offset = i * layer_spacing

    # Posiciones desplazadas
    pos = {n: (base_pos.get(n, (0, 0))[0] + offset,
               base_pos.get(n, (0, 0))[1])
           for n in G.nodes()}

    # Dibujar capa
    nx.draw(
        G,
        pos,
        ax=ax,
        node_size=20,
        alpha=0.6,
        with_labels=False
    )

    # Etiqueta de capa
    ax.text(offset, -1.5, season_key.upper(), ha="center")

    # Conexiones entre capas (mismo nodo)
    if i > 0:
        prev_offset = (i - 1) * layer_spacing
        for n in G.nodes():
            if n in graphs[f"s{i}"]:
                x1, y1 = base_pos.get(n, (0, 0))
                ax.plot(
                    [x1 + prev_offset, x1 + offset],
                    [y1, y1],
                    color="gray",
                    alpha=0.15,
                    linewidth=0.5
                )

ax.set_title("Multilayer network of Game of Thrones (Seasons 1–8)")
ax.axis("off")
plt.show()
```

**3.2 Entropía de Von Neumann**

Antes de la explicación con palabras propias, se presenta un breve resumen de los conceptos clave utilizados en los recursos de aprendizaje.

Según Domenico et al. (2015), la entropía de Von Neumann para un grafo se obtiene a partir de su matriz de Laplaciano normalizada.

La matriz de Laplaciano se define como:

$$
L = D - A
$$

donde **D** es la matriz diagonal de grados y **A** la matriz de adyacencia.


La normalización se obtiene dividiendo por la traza:

$$
\rho = \frac{L}{\mathrm{Tr}(L)}
$$


La entropía de Von Neumann **S** se calcula como:

$$
S = - \sum_i \lambda_i \, \log(\lambda_i)
$$

donde **$\lambda_i$** son los autovalores de la traza **$\rho$**.

Esta medida permite distinguir entre redes más complejas y diversas y aquellas más simples o redundantes. Valores altos indican mayor variedad estructural; valores bajos reflejan una organización más repetitiva.

Esta interpretación está alineada con la literatura en análisis de redes, donde la estructura y la distribución de conexiones son fundamentales (Newman, 2010).

Se ha utilizado el material de GitHub Docs (s.f.) – Writing mathematical expressions que explica 
cómo usar LaTeX en Markdown dentro de GitHub. https://docs.github.com/en/get-started/writing-on-github/working-with-advanced-formatting/writing-mathematical-expressions.

**3.2.1 Explicación en palabras propias**

En palabras sencillas, la entropía de Von Neumann mide cuán compleja o redundante es la estructura de una red. Para calcularla, se parte de la matriz de Laplaciano del grafo, que recoge la información de los nodos y sus conexiones. Esta matriz se normaliza dividiéndola por su traza, obteniendo una matriz de densidad. A partir de los autovalores de dicha matriz se calcula la entropía, que resume el grado de diversidad estructural de la red. Valores altos indican una red más compleja y diversa, mientras que valores bajos reflejan una estructura más simple o redundante.

Se calcula la entropía de Von Neumann para ser visualizada en el apartado 3.2.2:

```python
import networkx as nx
import numpy as np

def von_neumann_entropy(G):
    # Matriz de adyacencia
    A = nx.to_numpy_array(G)
    # Matriz diagonal de grados
    D = np.diag(A.sum(axis=1))
    # Laplaciano
    L = D - A

    # Normalización
    trace = np.trace(L)
    if trace == 0:
        return 0

    rho = L / trace

    # Autovalores
    eigvals = np.linalg.eigvalsh(rho)

    # Entropía
    eigvals = eigvals[eigvals > 0]  # filtrar ceros para evitar log(0)
    S = -np.sum(eigvals * np.log(eigvals))
    return S

# Ejemplo de uso
G = nx.erdos_renyi_graph(5, 0.5, seed=42)
entropy = von_neumann_entropy(G)
print("Entropía de Von Neumann:", entropy)
```

**Interpretación**

La entropía de Von Neumann mide la diversidad estructural de la red:

Valores altos - la red es más compleja y heterogénea, con nodos y conexiones variadas.

Valores bajos - la red es más simple o redundante, con estructura repetitiva o uniforme.

𝑆≈1.0986 Esto indica que la red tiene un nivel medio de diversidad estructural (S=0 sería un grafo totalmente desconectado, sin enlaces).

Con entropia de 1.0986 - hay cierta variedad, pero no es caos total; hay nodos con grados distintos, pero sin estructuras extremadamente dominantes.

**3.2.2 Aplicación a todas las temporadas y visualización**

En este ejercicio se utilizan los grafos cargados previamente en el Ejercicio 1. Se calcula la entropía de Von Neumann para cada temporada y se visualizan los resultados de manera comparativa.

```python
entropies = []

for season_key, G in graphs.items():
    entropies.append(von_neumann_entropy(G))

plt.figure(figsize=(8, 4))
plt.plot(range(1, 9), entropies, marker="o")
plt.xlabel("Temporada")
plt.ylabel("Entropía de Von Neumann")
plt.title("Entropía estructural por temporada")
plt.show()
```

**Modelo SCM sobre un RSC**

Definición del Random Simplicial Complex (RSC)

El RSC se construye a partir de:

Un grafo aleatorio de Erdős–Rényi, que representa interacciones binarias (aristas).

Un conjunto de triángulos, que representan interacciones de orden superior (2-símplices), añadidos sobre cliques de tamaño 3 con cierta probabilidad.

Esta construcción sigue la idea básica del artículo, permitiendo controlar la densidad de interacciones grupales.

**Simplicial Contagion Model (SCM)**

Cada nodo puede estar en dos estados:

- S (susceptible)

- I (infectado)

La dinámica incluye tres procesos:

Contagio binario
A través de aristas, con tasa $\lambda_1$

Contagio de orden superior
A través de triángulos, con tasa $\lambda_2$

Un nodo susceptible aumenta su probabilidad de infección si los otros dos nodos del triángulo están infectados.

Recuperación
Un nodo infectado se recupera con tasa $\mu$.

La probabilidad total de infección combina ambos mecanismos, reflejando la coexistencia de contagio individual y grupal.
Como señalan Lacopini et al. (2019), este tipo de modelo permite capturar sinergias colectivas imposibles de describir con grafos simples.

```python
**Ejercicio 1 – Implementación y comprobación del modelo** 

En este ejercicio se implementan dos bloques principales:

Generación del RSC (grafo Erdős–Rényi + triángulos).

Dinámica SCM con contagio binario, contagio de orden superior y recuperación.

(Este código contiene funciones que se llamarán después para reproducir la Figura 3 en el Ejercicio 2; aquí solo verificamos que el modelo funciona)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import random

# Random Simplicial Complex

def generate_rsc(N=300, p_edge=0.02, p_triangle=0.01, seed=42):
    random.seed(seed)
    np.random.seed(seed)

    # Grafo binario
    G = nx.erdos_renyi_graph(N, p_edge, seed=seed)

    # Triángulos (2-símplices)
    triangles = []
    nodes = list(G.nodes())

    for i in range(len(nodes)):
        for j in range(i+1, len(nodes)):
            for k in range(j+1, len(nodes)):
                u, v, w = nodes[i], nodes[j], nodes[k]
                if G.has_edge(u, v) and G.has_edge(v, w) and G.has_edge(u, w):
                    if random.random() < p_triangle:
                        triangles.append((u, v, w))

    return G, triangles


# Simplicial Contagion Model

def run_scm(G, triangles, lambda_1, lambda_2, mu, T, init_frac, seed=42):
    random.seed(seed)
    np.random.seed(seed)

    N = G.number_of_nodes()
    nodes = list(G.nodes())
    index = {n: i for i, n in enumerate(nodes)}

    # Estados: 0 = S, 1 = I
    state = np.zeros(N, dtype=int)
    state[random.sample(range(N), int(init_frac * N))] = 1

    # Vecinos
    neighbors = {n: [index[j] for j in G.neighbors(n)] for n in nodes}

    # Triángulos por nodo
    tri_by_node = {n: [] for n in nodes}
    for u, v, w in triangles:
        tri_by_node[u].append((index[v], index[w]))
        tri_by_node[v].append((index[u], index[w]))
        tri_by_node[w].append((index[u], index[v]))

    prevalence = []

    for _ in range(T):
        new_state = state.copy()

        # Recuperación
        new_state[(state == 1) & (np.random.rand(N) < mu)] = 0

        # Infección
        for i, n in enumerate(nodes):
            if state[i] == 0:
                k = np.sum(state[neighbors[n]])
                p1 = 1 - (1 - lambda_1)**k

                m = sum(1 for a, b in tri_by_node[n] if state[a] and state[b])
                p2 = 1 - (1 - lambda_2)**m

                p = 1 - (1 - p1)*(1 - p2)
                if np.random.rand() < p:
                    new_state[i] = 1

        state = new_state
        prevalence.append(np.mean(state))

    return np.array(prevalence)
```

**Verificación del modelo (simulación corta)**

Para comprobar el correcto funcionamiento del modelo implementado, se ejecuta una simulación corta del Simplicial Contagion Model sobre un Random Simplicial Complex.

En esta comprobación se imprimen:

- La prevalencia final del contagio.
- La prevalencia media en los últimos pasos temporales.

Esto permite verificar que:

- La dinámica del sistema evoluciona correctamente en el tiempo.
- El modelo no queda bloqueado en el estado inicial.
- El sistema alcanza un estado estacionario razonable, coherente con los parámetros del modelo.

Esta verificación es necesaria antes de utilizar el modelo para reproducir la Figura 3 del artículo en el Ejercicio 2.

```python
# Generar RSC
G, triangles = generate_rsc(N=300, p_edge=0.02, p_triangle=0.01)

print("Nodos:", G.number_of_nodes())
print("Aristas:", G.number_of_edges())
print("Triángulos:", len(triangles))

# Ejecutar simulación corta
prev = run_scm(
    G, triangles,
    lambda_1=0.4,
    lambda_2=1.2,
    mu=0.1,
    T=100,
    init_frac=0.1
)

print("Prevalencia final:", prev[-1])
print("Prevalencia media últimos pasos:", np.mean(prev[-20:]))

plt.plot(prev)
plt.xlabel("Tiempo")
plt.ylabel("Prevalencia")
plt.title("Verificación del SCM sobre un RSC")
plt.show()
```
